```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from sklearn.metrics import confusion_matrix
from tensorflow.keras.callbacks import ReduceLROnPlateau, ModelCheckpoint
```

```python
#Step 1:Dataset Selection
#Loading MNIST Dataset
(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()
```

```python
#Analyze Dataset
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
# Preprocess the data
x_train_reshaped = x_train.reshape(-1, 28, 28, 1).astype('float32') / 255.0
x_test_reshaped = x_test.reshape(-1, 28, 28, 1).astype('float32') / 255.0
```

```python
# Define a simpler CNN model
model = keras.Sequential([
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)),
    layers.MaxPooling2D((2, 2)),
    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dense(10, activation='softmax')
])
```

```python
# Compile the model with fixeed learning rate
optimizer = keras.optimizers.Adam(learning_rate=0.001)
model.compile(optimizer=optimizer , loss='sparse_categorical_crossentropy' , metrics=['accuracy'])
```

```python
# Apply learning rate decay strategies
# Define a learning rate scheduler
reduce_lr = ReduceLROnPlateau(monitor='val_accuracy' , factor=0.5 , patience=3 , min_lr=0.00001 , verbose=1)

# Train the model with the defined callback
history = model.fit(x_train_reshaped, y_train , epochs=3 , validation_data=(x_test_reshaped, y_test) , callbacks=[reduce_lr])
```
