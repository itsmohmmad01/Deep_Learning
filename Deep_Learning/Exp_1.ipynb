```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from sklearn.metrics import confusion_matrix
```

```python
#Step 1:Dataset Selection
#Loading MNIST Dataset
(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()
```

```python
#Analyze Dataset
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
#Step 2:Data Preprocessing
x_train=x_train.reshape(-1,784).astype("float32")/255.0
x_test=x_test.reshape(-1,784).astype("float32")/255.0
print(x_train.shape)
print(x_test.shape)

```

```python
#create simple neural network of 10 layers

#initalize ann
ann = tf.keras.models.Sequential()
```

```python
#Adding the input layer and the first hidden layer for 10 layers
ann.add(tf.keras.layers.Dense(units=6, activation='relu'))
ann.add(tf.keras.layers.Dense(units=6, activation='relu'))
ann.add(tf.keras.layers.Dense(units=6, activation='relu'))
ann.add(tf.keras.layers.Dense(units=6, activation='relu'))
ann.add(tf.keras.layers.Dense(units=6, activation='relu'))
ann.add(tf.keras.layers.Dense(units=6, activation='relu'))
ann.add(tf.keras.layers.Dense(units=6, activation='relu'))
ann.add(tf.keras.layers.Dense(units=6, activation='relu'))
ann.add(tf.keras.layers.Dense(units=6, activation='relu'))
ann.add(tf.keras.layers.Dense(units=6, activation='relu'))
```

```python
#Adding the output layer
ann.add(tf.keras.layers.Dense(units=10, activation='softmax'))
```

```python
#For : adam
#training model
ann.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['acc'])

#Compile ann
ann.fit(x_train,y_train,batch_size=32,epochs=5)

#Model Evalution
test_loss, test_acc = ann.evaluate(x_test, y_test)
print('Test accuracy:', test_acc)


```

```python
#For : SGD
#training model
ann.compile(optimizer='sgd',loss='sparse_categorical_crossentropy',metrics=['accuracy'])

#Compile ann
ann.fit(x_train,y_train,batch_size=32,epochs=10)

#Model Evalution
test_loss, test_acc = ann.evaluate(x_test, y_test)
print('Test accuracy:', test_acc)
```
